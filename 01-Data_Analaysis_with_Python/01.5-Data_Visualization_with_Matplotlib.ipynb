# Data Visualization with Matplotlib
Matplotlib is a powerful library for creating static, animated, and interactive visualizations in Python. 

It is widely used for data visualization in the scientific and engineering communities.

```python
import matplotlib.pyplot as plt
```

## Line Plot

Line Plot is used to display information as a series of data points called 'markers' connected by straight line segments. It is a basic type of chart common in many fields.

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y)
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
## Customizing Line Plots

plt.plot(x, y, color='red', linestyle='--', marker='o')
plt.title("Customized Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid()
plt.show()
```

```python
## Multiple Plots

x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [1, 2, 3, 4, 5]

plt.figure(figsize=(12,9))

plt.subplot(2,2,1)
plt.plot(x,y1,color='green')
plt.title("Plot 1")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.subplot(2,2,2)
plt.plot(y1,x,color='red')
plt.title("Plot 2")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.subplot(2,2,3)
plt.plot(x,y2,color='blue')
plt.title("Plot 3")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.subplot(2,2,4)
plt.plot(x,y2,color='green')
plt.title("Plot 4")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

## Bar Plots

Bar Plots are used to display categorical data with rectangular bars. The length of the bar is proportional to the value it represents. Bar plots can be plotted vertically or horizontally.

```python
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 4, 6]

plt.bar(categories, values, color='orange')
plt.title("Bar Plot")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

## Histogram

Histogram is a representation of the distribution of numerical data, and it is created by dividing the data into bins of equal width and counting the number of data points in each bin.

```python
data = [1, 1, 1, 1, 2, 2, 3, 3, 3, 5, 5, 5, 5, 5, 4, 4, 4, 4]

plt.hist(data, bins=5, color='purple', alpha=0.7, edgecolor='black')
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

## Scatter Plots

Scatter Plots are used to display values for typically two variables for a set of data. They are useful for observing relationships between variables.

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 4, 5, 6]

plt.scatter(x, y, color='red', marker='x')
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

## Pie Chart

Pie Chart is a circular statistical graphic divided into slices to illustrate numerical proportions. Each slice represents a category's contribution to the whole.


```python
labels = ['A', 'B', 'C', 'D', 'E']
sizes = [15, 30, 45, 10, 5]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'violet']
explode = (0.1, 0, 0, 0.1, 0)  # explode 1st slice

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
```

## Sales Data Visualization

```python
import pandas as pd
```

```python
df = pd.read_csv('data/sales_data.csv')
df.head()
```

```python
df.info()
```

```python
total_sales = df.groupby('Product Category')['Total Revenue'].sum()
total_sales = total_sales.sort_values(ascending=False)
total_sales
```

```python
total_sales.plot(kind='bar', figsize=(12, 6))
```

```python
## plot sales trend over time
sales_trend = df.groupby('Date')['Total Revenue'].sum().reset_index()
plt.plot(sales_trend['Date'], sales_trend['Total Revenue'])
```
