# Data Visualization with Seaborn

Seaborn is built on top of Matplotlib and provides a high-level interface for drawing attractive statistical graphics.

It is particularly well-suited for visualizing complex datasets and making it easy to create informative plots.

```python
import seaborn as sns
```

## Basic Plotting with Seaborn

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
import matplotlib.pyplot as plt

sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title("Scatter Plot")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
## Customizing Line Plots

plt.plot(x, y, color='red', linestyle='--', marker='o')
plt.title("Customized Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid()
plt.show()
```

## Line Plot

```python
sns.lineplot(x="size", y="total_bill", data=tips)
plt.title("Line Plot")
plt.xlabel("Size")
plt.ylabel("Total Bill")
plt.show()
```

## Bar Plots

Bar Plots are used to display categorical data with rectangular bars. The length of the bar is proportional to the value it represents. Bar plots can be plotted vertically or horizontally.

```python
sns.barplot(x='day', y='total_bill', data=tips, hue="sex", palette="viridis")
plt.title("Bar Plot of Total Bill by Day and Sex")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.show()
```

## Box Plot

A box plot is a standardized way of displaying the distribution of data based on a five-number summary ("minimum", first quartile (Q1), median, third quartile (Q3), and "maximum").


```python
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title("Box Plot of Total Bill by Day")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.show()
```

## Violin Plot

A violin plot is a method of plotting numeric data and can be understood as a combination of a box plot and a kernel density plot.

```python
sns.violinplot(x='day', y='total_bill', data=tips)
plt.title("Violin Plot of Total Bill by Day")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.show()
```

## Histogram

A histogram is a representation of the distribution of numerical data, and it is created by dividing the data into bins of equal width and counting the number of data points in each bin.


```python
sns.histplot(data=tips, x='total_bill', bins=10, kde=True)
plt.title("Histogram of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()
```

## KDE Plot

A Kernel Density Estimate (KDE) plot is a way to visualize the distribution of a continuous variable. It smooths the observed data points to create a continuous probability density curve.


```python
sns.kdeplot(tips['total_bill'],fill=True)
```

## Pair Plot

A pair plot is a matrix of scatter plots that shows the relationships between multiple variables in a dataset. It is useful for visualizing the pairwise relationships and distributions of the variables.


```python
sns.pairplot(tips)
```

## Heatmap

A heatmap is a graphical representation of data where individual values are represented as colors. It is useful for visualizing correlations between variables.


```python
sns.heatmap(data=tips[['total_bill', 'tip', 'size']].corr(), annot=True, cmap='coolwarm')
plt.title("Heatmap of Tips Dataset")
plt.show()
```
